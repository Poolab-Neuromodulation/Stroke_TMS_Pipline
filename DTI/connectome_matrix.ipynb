```python
import os
import pandas as pd
import seaborn as sns
```

```python
root = f'/home/wy/MRI/DTI/AFQ/MRtrix_Analysis_Scripts-master/mrtrix_data/lmf_1/raw2'
connect_path = os.path.join(root, 'connectome_a2009s.csv')
connect_fa_path = os.path.join(root, 'connectome_fa.csv')
```

```python
connect_a2009s = pd.read_csv(connect_path, header=None)
ax = sns.heatmap(connect_a2009s,vmax=0.08)
ax.set_title('Volume')
ax.set_xlabel('brain areas')
ax.set_ylabel('brain areas')
ax.figure.savefig(os.path.join(root, 'connectome_wholebrain.png'),dpi=600)
```

```python
connect_fa = pd.read_csv(connect_fa_path, header=None)
ax1 = sns.heatmap(connect_fa)
ax1.set_title('FA')
ax1.set_xlabel('brain areas')
ax1.set_ylabel('brain areas')
ax1.figure.savefig(os.path.join(root, 'connectome_fa.png'),dpi=600)
```
